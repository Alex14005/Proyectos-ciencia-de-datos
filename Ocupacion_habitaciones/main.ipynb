```python
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, confusion_matrix, precision_score, pair_confusion_matrix, recall_score
from scipy.stats import randint
```

```python
x = df.drop('Room_Occupancy_Count', axis=1)
y = df['Room_Occupancy_Count']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)

#Aplicamos PCA para reducir la dimensionalidad
pca = PCA(n_components=0.95)
X_train_pca = pca.fit_transform(X_train_scaler)
X_test_pca = pca.transform(X_test_scaler)

rf_classifier = RandomForestClassifier(random_state=42)

#hiperparametros a explorar
param_dist = {
    'n_estimators': randint(50, 200),
    'max_depth': randint(5, 30),
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 10),
    'max_features': ['auto', 'sqrt']
}

#objeto RandomizedSearchCV
random_search = RandomizedSearchCV(rf_classifier, param_distributions=param_dist, n_iter=20, cv=5, scoring='accuracy', random_state=42)

random_search.fit(X_train_pca, y_train)
best_params = random_search.best_params_

#Se entrena el modelo con los mejores hiperparametros
best_rf_classifier = RandomForestClassifier(**best_params, random_state=42)
best_rf_classifier.fit(X_train_pca, y_train)

#Realiza la validacion cruzada con DecisionTreeClassifier
rf_scores = cross_val_score(best_rf_classifier, X_train_pca, y_train, cv=5, scoring='accuracy') 
```

```python
dt_classifier = DecisionTreeClassifier(random_state=42)

param_dist = {
    'criterion': ['gni', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': randint(5, 30),
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 10),
    'max_features': ['auto', 'sqrt']
}

#Se crea el objeto RandomizedSearchCV
random_search = RandomizedSearchCV(dt_classifier, param_distributions=param_dist, n_iter=20, cv=5, scoring='accuracy', random_state=42)

random_search.fit(X_train_pca, y_train)

#Se obtienen los mejores hiperparametros
best_params = random_search.best_params_

#Se entrena el modelo con los mejores hiperparametros
best_dt_classifier = DecisionTreeClassifier(**best_params, random_state=42)
best_dt_classifier.fit(X_train_pca, y_train)

#Realiza la validacion cruzada con DecisionTreeClassifier
dt_scores = cross_val_score(best_dt_classifier, X_train_pca, y_train, cv=5, scoring='accuracy')

```

```python
#Imprimir los resultados de la validacion cruzada
print('Resultados de la validacion cruzada')
print(rf_scores)
print('Promedio de precisión con Random Forest: ', rf_scores.mean())

print("\Resultados de la validacion cruzada Con Decision Tree")
print(dt_scores)
print('Promedio de precisión con Decision Tree: ', dt_scores.mean())
```

```python
#Hacer predicciones con Random Forest
rf_predictions = best_rf_classifier.predict(X_test_pca)

#Hacer predicciones con Decision Tree
dt_predictions = best_dt_classifier.predict(X_test_pca)

#Evaluar el rendimiento del modelo con Random Forest
rf_f1 = f1_score(y_test, rf_predictions, average='weighted')
rf_precision = precision_score(y_test, rf_predictions, average='weighted')
rf_recall = recall_score(y_test, rf_predictions, average='weighted')

print('Rendimiento del modelo con Random Forest')
print('f1 score: ', rf_f1)
print('Precision: ', rf_precision)
print('Recall: ', rf_recall)

#Evaluar el rendimiento del modelo con Decision Tree
dt_f1 = f1_score(y_test, dt_predictions, average='weighted')
dt_precision = precision_score(y_test, dt_predictions, average='weighted')
dt_recall = recall_score(y_test, dt_predictions, average='weighted')

print('\nRendimiento del modelo con Decision Tree')
print('f1 score: ', dt_f1)
print('Precision: ', dt_precision)
print('Recall: ', dt_recall)

#Matriz de confusión con Random Forest
rf_confusion_matrix = confusion_matrix(y_test, rf_predictions)
print(rf_confusion_matrix)

#Matriz de confusión con Decision Tree
dt_confusion_matrix = confusion_matrix(y_test, dt_predictions)
print(dt_confusion_matrix)
```
